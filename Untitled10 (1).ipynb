```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = (5,5)
```

```python
def remove_outliers(df):
  for col in df:
    c = df[col]
    Q1 = np.percentile(c, 25, interpolation = 'midpoint')
    Q2 = np.percentile(c, 50, interpolation = 'midpoint')  
    Q3 = np.percentile(c, 75, interpolation = 'midpoint')
    IQR = Q3 - Q1
    print('Interquartile range is', IQR)
    low_lim = Q1 - 1.5 * IQR
    up_lim = Q3 + 1.5 * IQR
    print('low_limit is', low_lim)
    print('up_limit is', up_lim)
    outlier =[]
    for index, row in df.iterrows():
        if ((row[col]> up_lim) or (row[col]<low_lim)):
          df = df.drop(index)
  return df
```

```python
df = pd.read_csv("/content/Data.csv")
fix = df.drop(2759)
fix = fix.drop(9394)
new_df = fix.dropna()
```

```python
new = new_df[['AVG Views','Time Streamed' ,'Hours Watched num', 'Followers gaind num','Total Followers num','Total Views num']].copy()
new.head(10)
```

```python
proccesed = remove_outliers(new.loc[:,["AVG Views","Time Streamed","Hours Watched num","Followers gaind num","Total Followers num","Total Views num"]]).copy()#Remove all of the outliers in the data with the 1.5IQR method
proccesed.head(10)
```

```python
lr = linear_model.LinearRegression()#Create the linear regression model
proccesed["TimeXHours"] = proccesed["Time Streamed"]*proccesed["Hours Watched num"] #Add TimeXHours feature because of high correlation
proccesed["ViewsXfollowers"] = proccesed["Total Followers num"]*proccesed["Total Followers num"] #Add ViewsXFollowers feature because of high correlation

x = proccesed[["Time Streamed","Hours Watched num","Followers gaind num","Total Followers num","Total Views num","TimeXHours","ViewsXfollowers"]].values.reshape(proccesed[["Time Streamed","Hours Watched num","Followers gaind num","Total Followers num","Total Views num","TimeXHours","ViewsXfollowers"]].shape[0],7)
y = proccesed['AVG Views'].values.reshape(proccesed['AVG Views'].shape[0],1)

lr.fit(X=x, y=y)
print("Slope:",lr.coef_)
print("Intercept:",lr.intercept_)
print("R2:",lr.score(x,y))
```
